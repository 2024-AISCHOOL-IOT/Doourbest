 # 교차검증
  - cross_val_score

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# 데이터 불러오기
file_path = '../data/example.csv'
data = pd.read_csv(file_path)

# 필요한 열 선택 (예: 팀 통계 및 결과)
features = data[['FG_PCT_home',
                 'FG_PCT_away']]
target = data['HOME_TEAM_WINS']

# 데이터를 학습용과 테스트용으로 분리
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# 랜덤 포레스트 모델 초기화
model = RandomForestClassifier(random_state=42)

# 교차 검증 수행
scores = cross_val_score(model, X_train, y_train, cv=5)

# 모델 학습
model.fit(X_train, y_train)

# 예측 수행
y_pred = model.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)

# 결과 출력
print(f"교차 검증 점수: {scores}")
print(f"평균 교차 검증 점수: {scores.mean()}")
print(f"테스트 세트 정확도: {accuracy}")
```

# 가장 높은 cross_val_score (test_size=0.3)

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 데이터 불러오기
file_path = '../data/example.csv'
data = pd.read_csv(file_path)

# 필요한 열 선택 (예: 팀 통계 및 결과)
features = data[['FG_PCT_home',
                 'FG_PCT_away']]
target = data['HOME_TEAM_WINS']

# 데이터를 학습용과 테스트용으로 분리
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# 랜덤 포레스트 모델 초기화
model = RandomForestClassifier(random_state=42)

# 교차 검증 수행
scores = cross_val_score(model, X_train, y_train, cv=5)

# 모델 학습
model.fit(X_train, y_train)

# 예측 수행
y_pred = model.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)

# 결과 출력
print(f"교차 검증 점수: {scores}")
print(f"평균 교차 검증 점수: {scores.mean()}")
print(f"테스트 세트 정확도: {accuracy}")
```

```python
# knn 모델 객체 생성 해보기(하이퍼 파라미터 기본값 유지)
knn_model = KNeighborsClassifier(n_neighbors=9)
# n_neighbors=5 이웃의 개수 5개를 확인하며 예측을 진행함
tree_model = DecisionTreeClassifier(random_state=5)
```

```python
# 학습
knn_model.fit(X_train, y_train) # 거리 계산 규칙
tree_model.fit(X_train,y_train) # 규칙
```

```python
# test 에측
knn_pre = knn_model.predict(X_test)
print('knn_pre', knn_pre)
tree_pre = tree_model.predict(X_test)
print('tree_pre:', tree_pre)
# 모델 마다 서로 다른 규칙을 생성하기에 예측결과는 다를 수 있음
# 또 모델 내부의 하이퍼 파라미터 값에 대한 영향을 받을 수 있음
```

```python
knn_acc = knn_model.score(X_test, y_test)
tree_acc = tree_model.score(X_test,y_test)
print('knn_acc:', knn_acc)
print('tree_acc:', tree_acc)
```
